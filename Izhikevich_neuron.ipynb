```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#1.Simulation Parameters
T=200  #total time(ms)
dt= 0.25  #time steps (ms)
time = np.arange(0,T ,dt)
```

```python
#2.Regular spiking neurons
a= 0.02
b= 0.2
c=-65
d=8
```

```python
#3.Initializing the variables
v = -65 *np.ones(len(time))   #membrane potential 
u = b*v                       #recovery variable

I = np.zeros(len(time))
I[50:150] =10                 #external current input
```

```python
#4 . Simulation Loop 
for t in range(1 ,len(time)):
    v[t] = v[t-1] +dt *(0.04*v[t-1]**2 + 5*v[t-1] + 140 -u[t-1] + I[t])
    u[t] = u[t-1] + dt *(a*(b*v[t-1] - u[t-1]))

    if v[t] >= 30:
      v[t-1] = 30                #spike shown at 30 mV
      v[t] = c                   #reset voltage
      u[t] = u[t] + d            #recovery jump 
```

```python
# 5. Plot the graph
plt.figure(figsize=(10,4))
plt.plot(time , v)
plt.xlabel("Time(ms)")
plt.ylabel("Membrane potential(mV)")
plt.title("Izhevich neuron model(regular spiking)")
plt.show()
```

```python
#Parameters for the paper 
a = 0.03
b = -2
c = -50
d = 100


# Initialize variables

v = -65 * np.ones(len(time))   # membrane potential
u = b * v                      # recovery variable

I = np.zeros(len(time))
I[50:150] = 10                 # external current input


# Simulation loop

for t in range(1, len(time)):
    v[t] = v[t-1] + dt * (0.04*v[t-1]**2 + 5*v[t-1] + 140 - u[t-1] + I[t])
    u[t] = u[t-1] + dt * (a*(b*v[t-1] - u[t-1]))

    if v[t] >= 30:
        v[t-1] = 30           # spike shown at 30 mV
        v[t] = c              # reset voltage
        u[t] = u[t] + d       # recovery jump


# Plot results

plt.figure(figsize=(10,4))
plt.plot(time, v)
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.title("Izhikevich Neuron Model (Regular Spiking)")
plt.show()
```

# Observation:
After changing the parameters of the a , b, c, d which represent how fast the neuron recovers after firing, sensitivity of the recovery to the voltage , reset voltage after spike and recovery jump after spike we observed that there is the change in the graph of the normal vs the new valued graph.

## Regular spiking
The change occured because the first graph shows the steady predictable "drumbeat" that slows down.
## Fast/Bursting Variant 
The graph is much more "violent"-the reset points are higher , but the recovery jumps are so large that the neuron likely stuggles to maintain a steady firing rate without very high input current
