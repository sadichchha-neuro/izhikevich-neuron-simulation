```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
T = 200
dt = 0.5
time = np.arange(0,T,dt)
```

```python
print(np)
```

```python
print(time)
```

```python
a = 0.02 #how fast the neuron recoveres after firing
b = 0.2 #sensitivity of recovery to voltage
c = -65 #reset voltage after spike
d = 8 #recovery jump after spike
I = 15
v = -65 #initial membrane potential
u = b*v #initial recovery variable
```

```python
v_trace=[]
```

```python
print("initial u",u)
print("initial v",v)
```

```python
for t in time:
    #update the equation
    dv = (0.04*v**2 + 5*v + 140-u + I)*dt
    du = (a*(b*v-u))*dt

    v +=dv
    u +=du

    #spike condition
    if v>=30:
        v_trace.append(30) #record spike
        v = c
        u += d
    else:
        v_trace.append(c)
         
```

```python
len(v_trace)
```

```python
len(time)
```

```python
v_trace = np.array(v_trace)
```

```python
plt.figure(figsize=(12,4))
plt.plot(time, v_trace ,color='black')
plt.xlabel("Time(ms)")
plt.ylabel("Membrane potential(mV)")
plt.title("Izhikevich Neuron: Regular Spiking Behavior")
plt.show()
```
